```python
data.head(5)
```

```python
# relabel some of the data for descriptive legend, takes a min b/c for loop
for i in range(len(data.stableVar)):
    if data.loc[i,'stableVar'] == 0:
        data.loc[i,'stableVar'] = '$\sigma_s = \sigma_r$'
    elif data.loc[i,'stableVar'] == 1:
        data.loc[i,'stableVar'] = '$\sigma_s = 2\sigma_r$'
    elif data.loc[i,'stableVar'] == 2:
        data.loc[i,'stableVar'] = '$\sigma_s = 4\sigma_r$'
    elif data.loc[i,'stableVar'] == 3:
        data.loc[i,'stableVar'] = '$\sigma_s = 6\sigma_r$'
    if data.loc[i,'rwd_pen'] == 0:
        data.loc[i,'rwd_pen'] = '1:-2'
    elif data.loc[i,'rwd_pen'] == 1:
        data.loc[i,'rwd_pen'] = '1:-1'
    elif data.loc[i,'rwd_pen'] == 2:
        data.loc[i,'rwd_pen'] = '2:-1'
    elif data.loc[i,'rwd_pen'] == 3:
        data.loc[i,'rwd_pen'] = '4:-1'
```

```python
# compute z-scores for each subject's slxn err and peak velocity based on the response
def z_score(df):
    cols = ['slxn_err','peak_V']
    for col in cols:
        z_col = col + '_z_score'
        df[z_col] = (df[col] - df[col].mean())/df[col].std()
    return df

data = data.groupby(['subj_idx','response']).apply(z_score)
mask = (data.response == 0)
data.loc[[42, 314, 2017, 3005, 9001 ],:] # look at some random data points to check changes
```

## Choices 

```python
fig = plt.figure()
ax = sns.barplot(x='rwd_pen', y='response', hue='stableVar', 
                 order=['1:-2','1:-1','2:-1','4:-1'],
                 hue_order=['$\sigma_s = \sigma_r$', '$\sigma_s = 2\sigma_r$',
                            '$\sigma_s = 4\sigma_r$', '$\sigma_s = 6\sigma_r$'],
                 data=data, palette='Greys_d', saturation=.95)
ax.set_xlabel('Reward-to-Penalty Ratio (Risky Target)', fontsize=22)
ax.set_ylabel('p(Choice = Risky)', fontsize=22)
ax.set_yticklabels(np.linspace(0,1, 6), fontsize=20)
ax.set_ylim(top=1)
ax.set_xticklabels(['1:-2','1:-1','2:-1','4:-1'], fontsize=20)
ax.legend(loc='best',fontsize=22)
sns.despine()

# Uncomment below to save
fig.savefig('exp1choices.svg')
fig.savefig('exp1choices.eps')
# fig.savefig('exp1choices.png')
```
