The main purpose of this script is to visualize & preprocess the behavioral data prior to running analysis. Primarily, the removal of extreme outliers & subjects giving indications that they have genuinely attempted the task will be determined in these steps

## Import & Set Up 

```python
# Import packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

# enough with the warnings! ~ Thanks, Kyle
warnings.simplefilter('ignore', np.RankWarning)
warnings.filterwarnings("ignore", module="matplotlib")
warnings.filterwarnings("ignore", module="statsmodels")

# PLOT STYLING
# Display plots in the notebook
%matplotlib inline 

#if you have jupyterthemes:
from jupyterthemes import jtplot
jtplot.style('grade3',context='paper', fscale=1.3)
jtplot.figsize(6, 6, 1.2)
# if you don't have jupyterthemes, but have seaborn you can use:
# sns.set(context='paper', rc={'figure.facecolor':'white'}, font_scale=1.7)
```

## Import Data

```python
data_path = 'C:/Users/Alex Asturias/Documents/GitHub/Net_Worth/Data/' # Define the path to the behavioral data
data1 = pd.read_csv(data_path + 'exp1data.csv')
data2 = pd.read_csv(data_path + 'exp2data.csv')
data3 = pd.read_csv(data_path + 'exp3data.csv')
```

## Experiment 1

```python
data1.head(5)
# Convert the peak velocities into pixels/sec (currently pixels/ms)
data1['peak_V'] = data1['peak_V']*1000
data1.head(5)
```

### Reaction Times

#### Individual RT's

##### Trial-wise 

```python
fig,ax = plt.subplots(1,1);
data1.groupby('subj_idx').rt.plot();
ax.set_ylim([0,6]);
ax.set_title('RT by Subject');
ax.set_xlabel('trial');
ax.set_ylabel('RT (sec)');
# We see a couple of very slow RT's (on the order of seconds), but nothing alarming
```

##### Mean/Median 

```python
fig,(ax1, ax2) = plt.subplots(1,2,sharey = True);
data1.groupby('subj_idx').rt.mean().plot(ax=ax1);
ax1.set_title('Mean')
data1.groupby('subj_idx').rt.median().plot(ax=ax2);
ax2.set_title('Median')
ax1.set_ylabel('RT (sec)')
```

### Peak Velocity

#### Individual-level

##### Trial-wise

```python
fig,ax = plt.subplots(1,1);
data1.groupby('subj_idx').peak_V.plot();
plt.legend();
ax.set_ylim([0,6]);
ax.set_title('Peak V by Subject');
ax.set_xlabel('trial');
ax.set_ylabel('PeakV (pixels/ms)');
# May want to investigate the method for calculating the velocities, since these values seem low compared with the other tasks

# Clear that subject 9 is moving too rapidly, so we will remove them:
data1 = data1[data1['subj_idx'] != 9]
fig,ax = plt.subplots(1,1);
data1.groupby('subj_idx').peak_V.plot();
plt.legend();
ax.set_title('Peak V by Subject');
ax.set_xlabel('trial');
ax.set_ylabel('PeakV (pixels/ms)');
```

##### Mean/Median

```python
fig,(ax1, ax2) = plt.subplots(1,2,sharey = True);
data1.groupby('subj_idx').peak_V.mean().plot(ax=ax1);
ax1.set_title('Mean')
data1.groupby('subj_idx').peak_V.median().plot(ax=ax2);
ax2.set_title('Median')
ax1.set_ylabel('Peak Velocity (pixels/sec)');
```

### Selection Error

#### Individual-level

##### Trial-wise

```python
fig,ax = plt.subplots(1,1);
data1.groupby('subj_idx').slxn_err.plot();
ax.set_title('Slxn Err by Subject');
ax.set_xlabel('trial');
ax.set_ylabel('Error (pixels)');
# plt.legend();
```

##### Mean/Median

```python
fig,(ax1, ax2) = plt.subplots(1,2,sharey = True);
data1.groupby('subj_idx').slxn_err.mean().plot(ax=ax1);
ax1.set_title('Mean')
data1.groupby('subj_idx').slxn_err.median().plot(ax=ax2);
ax2.set_title('Median')
ax1.set_ylabel('Error (pixels)');
```
