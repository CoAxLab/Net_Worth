# Experiment 3 Behavioral Results Figures

## Imports & Set Up

```python
# Import packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

# enough with the warnings!
warnings.simplefilter('ignore', np.RankWarning)
warnings.filterwarnings("ignore", module="matplotlib")
warnings.filterwarnings("ignore", module="statsmodels")

# PLOT STYLING
# Display plots in the notebook
%matplotlib inline 

#if you have jupyterthemes:
from jupyterthemes import jtplot
jtplot.style('grade3',context='paper', fscale=1.3)
jtplot.figsize(6, 6, 1.2)
# if you don't have jupyterthemes, but have seaborn you can use:
# sns.set(context='paper', rc={'figure.facecolor':'white'}, font_scale=1.7)
```

## Data Manipulation

```python
data= pd.read_csv('D:/Rory/CoAx/Git/Pub_Code_master/Data/exp3data.csv')
```

```python
data.head(5)
```

```python
# relabel some of the data for descriptive legend, takes a min b/c for loop
for i in range(len(data.stableVar)):
    if data.loc[i,'stableVar'] == 0:
        data.loc[i,'stableVar'] = '$\sigma_s = \sigma_r$'
    elif data.loc[i,'stableVar'] == 1:
        data.loc[i,'stableVar'] = '$\sigma_s = 2\sigma_r$'
    elif data.loc[i,'stableVar'] == 2:
        data.loc[i,'stableVar'] = '$\sigma_s = 4\sigma_r$'
    elif data.loc[i,'stableVar'] == 3:
        data.loc[i,'stableVar'] = '$\sigma_s = 6\sigma_r$'
    if data.loc[i,'rwd_prob'] == 0:
        data.loc[i,'rwd_prob'] = '.15'
    elif data.loc[i,'rwd_prob'] == 1:
        data.loc[i,'rwd_prob'] = '.3'
    elif data.loc[i,'rwd_prob'] == 2:
        data.loc[i,'rwd_prob'] = '.7'
    elif data.loc[i,'rwd_prob'] == 3:
        data.loc[i,'rwd_prob'] = '.85'
```

```python
# compute z-scores for each subject's rt, slxn err and peak v based on the response
def z_score(df):
    cols = ['slxn_err','peak_V']
    for col in cols:
        z_col = col + '_z_score'
        df[z_col] = (df[col] - df[col].mean())/df[col].std()
    return df

data = data.groupby(['subj_idx','response']).apply(z_score)
mask = (data.response == 0)
data.loc[[42, 314, 2017, 3005, 9001 ],:] # look at some random data points to see changes
```

## Choices

```python
fig = plt.figure()
ax = sns.barplot(x='rwd_prob', y='response', hue='stableVar', 
                 order=['.15','.3','.7','.85'],
                 hue_order=['$\sigma_s = \sigma_r$', '$\sigma_s = 2\sigma_r$',
                            '$\sigma_s = 4\sigma_r$', '$\sigma_s = 6\sigma_r$'],
                 data=data, palette='Greys_d', saturation=.95)
ax.set_xlabel('P(reward | Choice = Risky)', fontsize=22)
ax.set_ylabel('p(Choice = Risky)', fontsize=22)
ax.set_yticklabels(np.linspace(0,1, 6), fontsize=20)
ax.set_ylim(top=1)
ax.set_xticklabels(['.15', '.3', '.7', '.85'], fontsize=20)
ax.legend(loc='best',fontsize=22)
sns.despine()

# Save
fig.savefig('Exp3Choices.eps')
fig.savefig('Exp3Choices.svg')
# fig.savefig('Exp3Choices_grey.png')
```
