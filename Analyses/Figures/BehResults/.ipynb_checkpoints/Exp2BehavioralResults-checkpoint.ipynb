# Experiment 2 Behavioral Results Figures

## Imports & Set Up

```python
# Import packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

# enough with the warnings!
warnings.simplefilter('ignore', np.RankWarning)
warnings.filterwarnings("ignore", module="matplotlib")
warnings.filterwarnings("ignore", module="statsmodels")

# PLOT STYLING
# Display plots in the notebook
%matplotlib inline 

#if you have jupyterthemes:
from jupyterthemes import jtplot
jtplot.style('grade3',context='paper', fscale=1.3)
jtplot.figsize(6, 6, 1.2)
# if you don't have jupyterthemes, but have seaborn you can use:
# sns.set(context='paper', rc={'figure.facecolor':'white'}, font_scale=1.7)
```

## Data Manipulation

```python
data= pd.read_csv('D:/Rory/CoAx/Git/Pub_Code_master/Data/exp2data.csv')
```

```python
data.head()
```

```python
# relabel some of the data for descriptive legend
for i in range(len(data.rwd_pen)):
    if data.loc[i,'rwd_pen'] == 1:
        data.loc[i,'rwd_pen'] = 'reward = penalty'
    elif data.loc[i,'rwd_pen'] == 2:
        data.loc[i,'rwd_pen'] = 'reward = 2(penalty)'
```

```python
# compute z-scores for each subject's slxn err and peak velocity based on the response
def z_score(df):
    cols = ['slxn_err','peak_V']
    for col in cols:
        z_col = col + '_z_score'
        df[z_col] = (df[col] - df[col].mean())/df[col].std()
    return df

data = data.groupby(['subj_idx','response']).apply(z_score)
mask = (data.response == 0)
data.loc[[42, 314, 2017, 3005, 9001 ],:] # look at some random data points to see the changes
```

## Choices

```python
fig = plt.figure()
ax = sns.barplot(x='rwd_prob', y='response', hue='rwd_pen', 
                 hue_order=['reward = penalty','reward = 2(penalty)'],
                 data=data, palette='Greys_d', saturation=.95)
ax.set_xlabel('P(reward | Choice = Risky)', fontsize=22)
ax.set_ylabel('P(Choice = Risky)', fontsize=22)
ax.set_yticklabels(np.linspace(0,1, 6), fontsize=20)
ax.set_ylim(top=1)
ax.set_xticklabels(['.15', '.3', '.7', '.85'], fontsize=20)
ax.legend(loc='best',fontsize=22)
sns.despine()

# Save
fig.savefig('exp2choices.eps')
fig.savefig('exp2choices.svg')
# fig.savefig('exp2choices_grey.png')
```

## Reaction Time

### Risky Target

```python
fig = plt.figure()
ax = sns.barplot(x='rwd_prob', y='rt', hue='rwd_pen', 
                 hue_order=['reward = penalty','reward = 2(penalty)'],
                 data=data[~mask], palette='Greys_d', saturation=.95)

ax.set_xlabel('P(reward | Choice = Risky)', fontsize=22)
ax.set_ylabel('Reaction Time (ms)', fontsize=22)
ax.set_yticks(np.linspace(.14,.28,8))
ax.set_yticklabels(np.linspace(140,280,8,dtype=int), fontsize=20)
ax.set_ylim(bottom = .14, top=.28)
ax.set_xticklabels(['.15','.3','.7','.85'], fontsize=20)
lgd = ax.legend(bbox_to_anchor=(1.1,1),fontsize=22)
sns.despine()

# Save
# fig.savefig('exp2_rt_risky_grey.png', bbox_extra_artists=(lgd,), bbox_inches='tight')
fig.savefig('exp2_rt_risky.eps', bbox_extra_artists=(lgd,), bbox_inches='tight')
fig.savefig('exp2_rt_risky.svg', bbox_extra_artists=(lgd,), bbox_inches='tight')
```
